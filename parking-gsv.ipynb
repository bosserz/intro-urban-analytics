```python
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import os
```

```python
def download_street_view_image(location, heading=0, pitch=0, fov=90):
    api_key = os.getenv("GSV_API")
    if not api_key:
        raise ValueError("Google Street View API key not found in environment variable 'GSV_API'")
    
    base_url = "https://maps.googleapis.com/maps/api/streetview"
    params = {
        "size": "640x640",
        "location": location,
        "heading": heading,
        "pitch": pitch,
        "fov": fov,
        "key": api_key
    }
    response = requests.get(base_url, params=params)
    img = Image.open(BytesIO(response.content))
    return img
```

```python
# test location
location = "33.824820,-84.185144"
street_view_img = download_street_view_image(location)
```
